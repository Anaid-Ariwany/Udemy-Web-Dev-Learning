# Bootstrap

## Introduction

- Bootstrap: a popular CSS framework for building responsive and mobile-first websites.
- Provides pre-designed components and utilities for faster web development.

- Key Features:
  - Responsive Grid System: Allows for flexible layouts that adapt to different screen sizes.
  - Pre-styled Components: Includes buttons, forms, modals, navbars, and more.
  - Customizable: Easily override default styles with custom CSS or Sass variables.
  - JavaScript Plugins: Offers interactive components like carousels, tooltips, and modals.
  - Cross-browser Compatibility: Ensures consistent appearance across different web browsers.

1. Bootstrap version 4.5: https://getbootstrap.com/docs/4.5/getting-started/introduction/ 

2. Bootstrap version 5.3: https://getbootstrap.com/docs/5.3/getting-started/introduction/

### Setting Up

- You can include Bootstrap in your project via CDN or by downloading the files.
- It's recommended to use the latest stable version for new projects.
- The best way to setup Bootstrap is to use the CDN links provided in the official documentation. Downloading the files is also an option, but using the CDN ensures you always have the latest version and reduces the load on your server.

## Components and Utilities

- Components: bootstrap provides a variety of pre-designed components that can be easily integrated into your web pages to enhance functionality and aesthetics. Examples:
  - Buttons
  - Forms
  - Navbars
  - Modals
  - Cards
  - Alerts

- Utilities: are helper classes that allow you to quickly style elements without writing custom CSS. Examples include:
  - Spacing (margins and padding)
  - Text alignment
  - Display properties
  - Flexbox utilities
  - Color utilities

## The Grid System

### Intro

- Bootstrap Grid System: a powerful layout system that uses a series of containers, rows, and columns to create **responsive** layouts.
- Based on a 12-column layout, allowing for flexible and customizable designs. Therefore, each row can be divided into up to 12 columns. This is done by specifying the number of columns an element should span using classes like `.col-4` (which spans 4 out of 12 columns).

### Responsive Behavior

- The Grid already has specified breakpoints for different screen sizes (e.g., `.col-sm-`, `.col-md-`, `.col-lg-`, `.col-xl-`). These breakpoints are used in specific elements to define how many columns they should span at different screen sizes. For example in a div:
  `<div class="col-12 col-md-6 col-lg-4">`
  
  This means:
  - On extra small to small screens (less than 768px), the div will span all 12 columns (full width).
  - On medium screens (768px and up), it will span 6 columns (half width).
  - On large screens (992px and up), it will span 4 columns (one-third width).

- You can also use flexbox utilities to create more complex layouts within the grid system. eg: `div class="row justify-content-center"` to center content within a column.
- The flexbox utilities can also be made responsive by adding breakpoints, e.g., `justify-content-md-center` to center content only on medium and larger screens.
